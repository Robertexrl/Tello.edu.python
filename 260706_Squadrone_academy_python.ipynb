### What is a Tello drone?

The Tello is a small, lightweight quadcopter (it has four propellers). It is suitable for flying indoors, it can perform flips and rotations, and it can be controlled with Python code rather than a handheld controller.

### How does the drone communicate with the computer?

The Tello creates its own Wi-Fi network. Your computer connects to that network, and your Python code sends short text commands (such as "takeoff" or "move up") to the drone over the connection. The drone reads each command and carries it out.

We use a Python library called `djitellopy` to make this straightforward. It provides simple commands like `tello.takeoff()` and `tello.land()`, so the underlying networking is handled for us.

### Before you begin

1. Turn on the Tello drone (the light will blink).
2. On your computer, connect to the drone's Wi-Fi network (the name usually starts with `TELLO-`).
3. Install the library by running this in a terminal: `pip install djitellopy`
4. Ensure there is clear space around the drone.

### About flight distances

These missions assume a clear flying area of roughly **200 x 200 x 200 cm**. After take-off the drone rises to approximately **50 cm**, which is the starting height for any mission that descends. Any distance in the code can be adjusted to match the space you have available.

> **SAFETY:** Keep hands and faces clear of the propellers. Fly only in open space. You can run `tello.land()` at any time to bring the drone down.

## Mission 2: Fly a Square

Next, fly the drone in a **square** pattern. To trace a square, the drone flies forward, turns 90 degrees, and repeats four times.

**Why 90 degrees?** A square has four corners, and at each corner the drone turns a quarter of a full circle. A quarter of 360 degrees is 90 degrees.

**Steps:**
1. Take off.
2. Fly forward, then turn right 90 degrees. Repeat four times, once per side.
3. Land.

````python
tello.takeoff()

# Side 1
tello.move_forward(150)
tello.rotate_clockwise(90)

# Side 2
tello.move_forward(150)
tello.rotate_clockwise(90)

# Side 3
tello.move_forward(150)
tello.rotate_clockwise(90)

# Side 4
tello.move_forward(150)
tello.rotate_clockwise(90)

tello.land()

print("Mission 2 complete.")
````

## Mission 3: Fly a Triangle

A triangle is harder than a square. A triangle has only **three** corners, so you might expect the turn to be 360 / 3 = 120 degrees. That is correct.

**The rule for turning angles:** To fly any closed shape, the drone turns a **total of 360 degrees** by the time it returns to the start.
- Square: 4 turns, so each turn is 360 / 4 = **90 degrees**.
- Triangle: 3 turns, so each turn is 360 / 3 = **120 degrees**.

**Steps:**
1. Take off.
2. Fly forward, then turn 120 degrees. Repeat three times.
3. Land.

````python
tello.takeoff()

# Side 1
tello.move_forward(150)
tello.rotate_clockwise(120)

# Side 2
tello.move_forward(150)
tello.rotate_clockwise(120)

# Side 3
tello.move_forward(150)
tello.rotate_clockwise(120)

tello.land()

print("Mission 3 complete.")
````

## Mission 4: The Figure-8 (Hourglass)

This is the most demanding mission at this rank. A figure-8 resembles the number 8, or an hourglass. The key idea is that the drone must **change its turning direction** halfway through.

In the first loop the drone turns **right** (clockwise). It then crosses over, and in the second loop it turns **left** (counter-clockwise). Reversing the turn direction is what makes the path cross itself and form the 8.

Each loop is a triangle (three turns of 120 degrees). Two triangles meeting in the middle produce the hourglass shape.

**Steps:**
1. Take off.
2. Fly the first triangle loop, turning **right** each time.
3. Fly the second triangle loop, turning **left** each time.
4. Land.

Note how much code this requires. Mission 6 introduces a shorter way to write repeated instructions.

````python
tello.takeoff()

# First loop of the 8 - turning RIGHT (clockwise)
tello.move_forward(120)
tello.rotate_clockwise(120)
tello.move_forward(120)
tello.rotate_clockwise(120)
tello.move_forward(120)
tello.rotate_clockwise(120)

# Second loop of the 8 - now turning LEFT (counter-clockwise)
tello.move_forward(120)
tello.rotate_counter_clockwise(120)
tello.move_forward(120)
tello.rotate_counter_clockwise(120)
tello.move_forward(120)
tello.rotate_counter_clockwise(120)

tello.land()

print("Mission 4 complete.")
````

## Mission 5: Flying in 3D with Coordinates

So far the drone has moved straight forward, up, or sideways. To fly **diagonally** - moving in more than one direction at once - we use **coordinates**.

You give the drone three numbers, **x**, **y**, and **z**, describing where to go relative to its current position:

| Letter | Direction | Positive number | Negative number |
|---|---|---|---|
| **x** | forward / backward | forward | backward |
| **y** | left / right | left | right |
| **z** | up / down | up | down |

The command is `tello.go_xyz_speed(x, y, z, speed)`. The final number, **speed**, is the flying speed in cm per second.

For example, `tello.go_xyz_speed(70, 0, 60, 40)` means: fly 70 cm forward and 60 cm up at the same time, at a speed of 40. This produces a smooth diagonal.

> **Note:** each of x, y, and z must be 0 or at least 20 - the drone cannot make moves smaller than 20 cm.

**Important about height:** after take-off the drone is at about 50 cm. Before flying a pattern that includes downward moves, climb to a safe height first so the drone does not descend into the ground. In this mission we climb to about 100 cm, then trace a vertical diamond that stays between roughly 40 cm and 160 cm.

**Steps:**
1. Take off and climb to about 100 cm.
2. Fly four diagonal moves to trace a diamond standing on its edge.
3. Land.

````python
tello.takeoff()
tello.move_up(50)                    # climb from ~50 cm to ~100 cm for headroom

tello.go_xyz_speed(70, 0, 60, 40)    # diagonal: forward and up   (height now ~160 cm)
tello.go_xyz_speed(70, 0, -60, 40)   # diagonal: forward and down (height now ~100 cm)
tello.go_xyz_speed(-70, 0, -60, 40)  # diagonal: back and down    (height now ~40 cm)
tello.go_xyz_speed(-70, 0, 60, 40)   # diagonal: back and up      (height now ~100 cm)

tello.land()

print("Mission 5 complete.")
````

## Mission 6: Using Loops

The square in Mission 2 required writing nearly the same two lines four times. A **loop** is a shorter way to repeat instructions.

A `for` loop runs code a set number of times. This example prints "Hello" four times:

```python
for i in range(4):
    print("Hello")
```

The `range(4)` means run four times, and `i` is a counter that takes the values 0, 1, 2, 3. You can name it anything, but `i` is the convention.

The loop below reduces the entire Mission 2 square to two flight lines.

**Steps:**
1. Take off.
2. Use a loop that repeats four times: fly forward, then turn 90 degrees.
3. Land.

````python
tello.takeoff()

# Repeat the same two moves 4 times to draw a square
for i in range(4):
    tello.move_forward(150)
    tello.rotate_clockwise(90)

tello.land()

print("Mission 6 complete.")
````

- `tello.flip_right()`

> **Note:** flips require significant power, so the battery must be above roughly 50 percent or the drone will not perform them. Ensure there is plenty of space, as the drone moves during a flip.

**Steps:**

2. Perform a forward flip, then a back flip.

## Mission 8: Controlling Speed

The flight speed is set with `tello.set_speed(speed)`, where speed ranges from 10 (slow) to 100 (fast) cm per second.

In this mission the drone flies the same distance twice - once slowly, once quickly - so you can compare.

**Steps:**
1. Take off.
2. Set a slow speed and fly forward.

## Mission 10: Securing the Connection

This mission covers an important topic: **security**. The Tello drone creates an **open Wi-Fi network**, which means it has no password by default. Anyone nearby with a computer could connect to the drone and control it without permission. This is a genuine **cybersecurity** risk.

In real systems, engineers use passwords and access codes so that only authorised users can operate a device. In this mission you will add a **launch-code check**. The drone will only take off if the correct code is entered; if the code is wrong, it stays on the ground.

This introduces two concepts:
- `input(...)` asks the user to type something.
- `if ... else ...` lets the program make a decision based on the input.

**Steps:**
1. Define a secret launch code.
2. Ask the operator to type the code.
3. If the code is correct, the drone flies. Otherwise, it stays grounded.

````python
# The secret launch code. In real systems, a password should never be shared.
secret_code = "SQUADRONE2026"

# Ask the operator to enter the code
attempt = input("Enter the launch code to unlock the drone: ")

if attempt == secret_code:
    print("Access granted. Launching.")
    tello.takeoff()
    tello.rotate_clockwise(360)
    tello.land()
else:
    print("Access denied. Incorrect code - the drone will not fly.")

print("Mission 10 complete.")
````

## Mission 11: Writing a Function

When the same routine is flown repeatedly, you can give it a name and reuse it. This is what a **function** is - a named set of instructions that you can run whenever you need it.

A function is created with the keyword `def` (short for "define"):

```python
def my_function():
    # instructions go here, indented
```

Once defined, it is run by writing its name followed by brackets, such as `my_function()`.

In this mission we place the Hello World routine into a function named `hello_drone`, then call it.

**Steps:**
1. Define a function named `hello_drone` that takes off, climbs, rotates, and lands.
2. Call the function to run it.

````python
def hello_drone():
    tello.takeoff()
    tello.move_up(80)
    tello.rotate_counter_clockwise(360)
    tello.land()

# Run the function
hello_drone()

print("Mission 11 complete.")
````

## Mission 12: Functions with Parameters

Functions become more powerful when you give them an **input value**, called a **parameter**. This lets one function behave differently each time it is called.

For example, a `fly_square` function can take a `size` parameter, so the same code can fly a small or large square:

```python
def fly_square(size):
    ...
```

Then `fly_square(50)` flies a small square and `fly_square(150)` flies a large one.

**Steps:**
1. Define a function `fly_square(size)` that uses a loop to fly a square of the given size.
2. Call it with a small size.
3. Optionally, call it again with a larger size.

````python
def fly_square(size):
    tello.takeoff()
    for i in range(4):
        tello.move_forward(size)
        tello.rotate_clockwise(90)
    tello.land()

# Fly a small square (80 cm sides)
fly_square(80)

# Try this as well - remove the # to fly a larger square:
# fly_square(150)

print("Mission 12 complete.")
````

## Mission 13: Combining Functions

Larger programs are built from small functions joined together. In this mission you will write three small functions and call them in order to form a complete routine.

## Mission 14: Decisions Based on Sensors

A well-designed program checks that it is safe to fly before taking off. Flying with a low battery can cause the drone to fall, so the program should decide for itself.

This mission reads the battery sensor and uses an `if` statement: if the battery is high enough, the drone flies; otherwise it stays grounded and warns the operator. This combines sensors (Mission 9) with decisions (Mission 10).

**Steps:**
1. Read the battery level.
2. If the battery is above 30 percent, fly. Otherwise, print a warning and do not fly.

````python
battery = tello.get_battery()
print("Battery level is", battery, "%")

if battery > 30:
    print("Battery is sufficient - cleared for take-off.")
    tello.takeoff()
    tello.move_up(80)
    tello.land()
else:
    print("Battery is too low. It is not safe to fly. Please charge the drone.")

print("Mission 14 complete.")
````

## Mission 15: A Launch Countdown

Launches are often preceded by a countdown. This can be built using a loop, the `time.sleep()` command (which pauses the program), and a `range` that counts **backwards**.

`range(5, 0, -1)` counts 5, 4, 3, 2, 1. The third number, `-1`, means decrease by one each step.

**Steps:**
1. Print a preparation message.
2. Loop from 5 down to 1, printing each number and pausing one second.
3. Launch.

````python
print("Prepare for launch.")

for count in range(5, 0, -1):
    print(count)
    time.sleep(1)        # pause for 1 second

print("Launching.")
tello.takeoff()
tello.move_up(60)
tello.land()

print("Mission 15 complete.")
````

## Mission 16: A Flight Sequence

In this mission you will program a longer sequence by combining moves, rotations, and flips inside a single function. This is like writing out an ordered list of steps for the drone to perform.

**Steps:**
1. Define a `sequence` function with a series of moves.
2. Call it to run the sequence.

You can change the moves to create your own sequence.

````python
def sequence():
    tello.takeoff()
    tello.move_up(70)
    tello.rotate_clockwise(180)
    tello.rotate_counter_clockwise(180)
    tello.flip_left()
    tello.flip_right()
    tello.rotate_clockwise(360)
    tello.land()

sequence()

print("Mission 16 complete.")
````

## Mission 17: Random Flight

Using random numbers, the drone can fly a different path each time the code runs. Python's `random` library provides these numbers.

`random.randint(20, 80)` returns a whole number somewhere between 20 and 80.

**Steps:**
1. Import the random library.
2. Take off.
3. Loop a few times, each time flying a random distance and turning a random angle.
4. Land.

````python
import random

tello.takeoff()

for i in range(4):
    distance = random.randint(30, 80)    # random distance
    angle = random.randint(30, 180)      # random turn
    print("Flying", distance, "cm then turning", angle, "degrees")
    tello.move_forward(distance)
    tello.rotate_clockwise(angle)

tello.land()

print("Mission 17 complete.")
````

## Mission 18: A Combined Display

This mission brings together functions, loops, and a range of moves in one longer flight. It reuses the square idea, adds rotations, and finishes with a flip.

**Steps:**
1. Define a couple of helper functions for parts of the display.
2. Run them in sequence, using a loop for the repeating square pattern.

````python
def double_spin():
    for i in range(2):
        tello.rotate_clockwise(360)

def square_lap(size):
    for i in range(4):
        tello.move_forward(size)
        tello.rotate_clockwise(90)

# The combined display
tello.takeoff()
tello.move_up(70)
square_lap(100)       # fly a square
double_spin()         # rotate twice
tello.flip_forward()  # finish with a flip
tello.land()

print("Mission 18 complete.")
````

## Mission 19: Handling Errors Safely

A reliable program plans for things going wrong. A command can sometimes fail - for example, if the battery drops or the connection is interrupted. The flight can be protected using `try` and `except`.

- The code inside `try` is what should happen normally.
- If anything goes wrong, the code inside `except` runs instead - and we use it to land the drone safely regardless of the error.

**Steps:**
1. Place the flight inside a `try` block.
2. In the `except` block, print the problem and land safely.

````python
try:
    tello.takeoff()
    tello.move_up(80)
    tello.flip_forward()
    tello.land()
except Exception as e:
    print("Something went wrong:", e)
    print("Running emergency landing.")
    tello.land()

print("Mission 19 complete.")
````
