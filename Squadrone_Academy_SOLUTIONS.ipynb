### What is a Tello drone?

The Tello is a small, lightweight quadcopter (it has four propellers). It is suitable for flying indoors, it can perform flips and rotations, and it can be controlled with Python code rather than a handheld controller.

### How does the drone communicate with the computer?

The Tello creates its own Wi-Fi network. Your computer connects to that network, and your Python code sends short text commands (such as "takeoff" or "move up") to the drone over the connection. The drone reads each command and carries it out.

We use a Python library called `djitellopy` to make this straightforward. It provides simple commands like `tello.takeoff()` and `tello.land()`, so the underlying networking is handled for us.

### Before you begin

1. Turn on the Tello drone (the light will blink).
2. On your computer, connect to the drone's Wi-Fi network (the name usually starts with `TELLO-`).
3. Install the library by running this in a terminal: `pip install djitellopy`
4. Ensure there is clear space around the drone.

### Sharing one drone between pilots

In the Academy, several pilots share one drone. The drone accepts **only one computer at a time**, so every mission follows the same routine:

1. **Connect** your laptop to the drone's `TELLO-` Wi-Fi network.
2. **Run your mission.** Every mission begins with a PRE-FLIGHT block that connects to the drone, and ends with a POST-FLIGHT block that releases it.
3. When you see `Drone released - the next pilot can now connect.`, **disconnect** your laptop from the `TELLO-` Wi-Fi network.
4. The next pilot can now connect their laptop and fly their mission.

If a laptop stays connected to the drone's Wi-Fi, the next pilot will not be able to connect - so always complete step 3.

### About flight distances

These missions assume a clear flying area of roughly **200 x 200 x 200 cm**. After take-off the drone rises to approximately **50 cm**, which is the starting height for any mission that descends. Any distance in the code can be adjusted to match the space you have available.

### Flight commands

| Command | What it does | Allowed values |
|---|---|---|
| `tello.takeoff()` | lifts off and hovers at about 50 cm | - |
| `tello.land()` | lands gently | - |
| `tello.move_up(d)` / `tello.move_down(d)` | climbs / descends `d` cm | 20 to 500 |
| `tello.move_forward(d)` / `tello.move_back(d)` | flies forward / backward `d` cm | 20 to 500 |
| `tello.move_left(d)` / `tello.move_right(d)` | flies sideways `d` cm | 20 to 500 |
| `tello.rotate_clockwise(a)` | turns right by `a` degrees | 1 to 360 |
| `tello.rotate_counter_clockwise(a)` | turns left by `a` degrees | 1 to 360 |
| `tello.flip_forward()` / `_back()` / `_left()` / `_right()` | performs a flip | battery above about 50% |
| `tello.go_xyz_speed(x, y, z, speed)` | diagonal move: x forward, y left, z up (negative numbers reverse each) | each of x, y, z is 0 or between 20 and 500; speed 10 to 100 |
| `tello.set_speed(s)` | sets flying speed in cm per second | 10 to 100 |

### Sensor commands

| Command | What it reports |
|---|---|
| `tello.get_battery()` | remaining battery (percent) |
| `tello.get_height()` | current height (cm) |
| `tello.get_temperature()` | drone temperature (degrees C) |
| `tello.get_flight_time()` | time spent flying (seconds) |

### Python you will use

| Pattern | What it does |
|---|---|
| `for i in range(4):` | repeats the indented lines below it 4 times |
| `def my_function():` | defines a function you can run later with `my_function()` |
| `if a == b: ... else: ...` | makes a decision (`==` compares two values; a single `=` stores a value) |
| `answer = input("...")` | asks the operator to type something |
| `time.sleep(1)` | pauses for 1 second |
| `random.randint(20, 80)` | a random whole number between 20 and 80 (needs `import random`) |

> **A note about spaces:** Python is fussy about indentation. Every line inside a loop, a function, or an `if` must be indented by exactly four spaces. If Python reports an `IndentationError`, check the spaces at the start of each line.

## Mission 1: Hello World

A programmer's first program is traditionally called "Hello World". The drone equivalent is a short flight: the drone takes off, rises, rotates a full circle as a greeting, and lands.

## Mission 2: Fly a Square

Next, fly the drone in a **square** pattern. To trace a square, the drone flies forward, turns 90 degrees, and repeats four times.

**Why 90 degrees?** A square has four corners, and at each corner the drone turns a quarter of a full circle. A quarter of 360 degrees is 90 degrees.

**Steps:**
1. Take off.
2. Fly forward, then turn right 90 degrees. Repeat four times, once per side.
3. Land.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
tello.takeoff()

# Side 1
tello.move_forward(150)
tello.rotate_clockwise(90)

# Side 2
tello.move_forward(150)
tello.rotate_clockwise(90)

# Side 3
tello.move_forward(150)
tello.rotate_clockwise(90)

# Side 4
tello.move_forward(150)
tello.rotate_clockwise(90)

tello.land()

print("Mission 2 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 3: Fly a Triangle

A triangle is harder than a square. A triangle has only **three** corners, so you might expect the turn to be 360 / 3 = 120 degrees. That is correct.

**The rule for turning angles:** To fly any closed shape, the drone turns a **total of 360 degrees** by the time it returns to the start.
- Square: 4 turns, so each turn is 360 / 4 = **90 degrees**.
- Triangle: 3 turns, so each turn is 360 / 3 = **120 degrees**.

**Steps:**
1. Take off.
2. Fly forward, then turn 120 degrees. Repeat three times.
3. Land.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
tello.takeoff()

# Side 1
tello.move_forward(150)
tello.rotate_clockwise(120)

# Side 2
tello.move_forward(150)
tello.rotate_clockwise(120)

# Side 3
tello.move_forward(150)
tello.rotate_clockwise(120)

tello.land()

print("Mission 3 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 4: The Figure-8 (Hourglass)

This is the most demanding mission at this rank. A figure-8 resembles the number 8, or an hourglass. The key idea is that the drone must **change its turning direction** halfway through.

In the first loop the drone turns **right** (clockwise). It then crosses over, and in the second loop it turns **left** (counter-clockwise). Reversing the turn direction is what makes the path cross itself and form the 8.

Each loop is a triangle (three turns of 120 degrees). Two triangles meeting in the middle produce the hourglass shape.

**Steps:**
1. Take off.
2. Fly the first triangle loop, turning **right** each time.
3. Fly the second triangle loop, turning **left** each time.
4. Land.

Note how much code this requires. Mission 6 introduces a shorter way to write repeated instructions.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
tello.takeoff()

# First loop of the 8 - turning RIGHT (clockwise)
tello.move_forward(120)
tello.rotate_clockwise(120)
tello.move_forward(120)
tello.rotate_clockwise(120)
tello.move_forward(120)
tello.rotate_clockwise(120)

# Second loop of the 8 - now turning LEFT (counter-clockwise)
tello.move_forward(120)
tello.rotate_counter_clockwise(120)
tello.move_forward(120)
tello.rotate_counter_clockwise(120)
tello.move_forward(120)
tello.rotate_counter_clockwise(120)

tello.land()

print("Mission 4 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 5: Flying in 3D with Coordinates

So far the drone has moved straight forward, up, or sideways. To fly **diagonally** - moving in more than one direction at once - we use **coordinates**.

You give the drone three numbers, **x**, **y**, and **z**, describing where to go relative to its current position:

| Letter | Direction | Positive number | Negative number |
|---|---|---|---|
| **x** | forward / backward | forward | backward |
| **y** | left / right | left | right |
| **z** | up / down | up | down |

The command is `tello.go_xyz_speed(x, y, z, speed)`. The final number, **speed**, is the flying speed in cm per second.

For example, `tello.go_xyz_speed(70, 0, 60, 40)` means: fly 70 cm forward and 60 cm up at the same time, at a speed of 40. This produces a smooth diagonal.

> **Note:** each of x, y, and z must be 0 or at least 20 - the drone cannot make moves smaller than 20 cm.

**Important about height:** after take-off the drone is at about 50 cm. Before flying a pattern that includes downward moves, climb to a safe height first so the drone does not descend into the ground. In this mission we climb to about 100 cm, then trace a vertical diamond that stays between roughly 40 cm and 160 cm.

**Steps:**
1. Take off and climb to about 100 cm.
2. Fly four diagonal moves to trace a diamond standing on its edge.
3. Land.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
tello.takeoff()
tello.move_up(50)                    # climb from ~50 cm to ~100 cm for headroom

tello.go_xyz_speed(70, 0, 60, 40)    # diagonal: forward and up   (height now ~160 cm)
tello.go_xyz_speed(70, 0, -60, 40)   # diagonal: forward and down (height now ~100 cm)
tello.go_xyz_speed(-70, 0, -60, 40)  # diagonal: back and down    (height now ~40 cm)
tello.go_xyz_speed(-70, 0, 60, 40)   # diagonal: back and up      (height now ~100 cm)

tello.land()

print("Mission 5 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 6: Using Loops

The square in Mission 2 required writing nearly the same two lines four times. A **loop** is a shorter way to repeat instructions.

A `for` loop runs code a set number of times. This example prints "Hello" four times:

```python
for i in range(4):
    print("Hello")
```

The `range(4)` means run four times, and `i` is a counter that takes the values 0, 1, 2, 3. You can name it anything, but `i` is the convention.

The loop below reduces the entire Mission 2 square to two flight lines.

**Steps:**
1. Take off.
2. Use a loop that repeats four times: fly forward, then turn 90 degrees.
3. Land.

> **Watch your spaces:** the two flight lines inside the loop must each be indented by four spaces, or Python will report an `IndentationError`.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
tello.takeoff()

# Repeat the same two moves 4 times to draw a square
for i in range(4):
    tello.move_forward(150)
    tello.rotate_clockwise(90)

tello.land()

print("Mission 6 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 10: Securing the Connection

This mission covers an important topic: **security**. The Tello drone creates an **open Wi-Fi network**, which means it has no password by default. Anyone nearby with a computer could connect to the drone and control it without permission. This is a genuine **cybersecurity** risk.

In real systems, engineers use passwords and access codes so that only authorised users can operate a device. In this mission you will add a **launch-code check**. The drone will only take off if the correct code is entered; if the code is wrong, it stays on the ground.

This introduces two concepts: asking for input, and making a decision.

`input(...)` pauses the program and asks the operator to type something. Whatever they type is stored in a variable:

```python
attempt = input("Enter the launch code to unlock the drone: ")
```

An `if ... else ...` statement lets the program make a decision. It looks like this:

```python
if attempt == secret_code:
    # this runs when the answer is YES
else:
    # this runs when the answer is NO
```

Three things to notice:
- The double equals sign `==` asks a question: "are these two things the same?" A single `=` does something different - it stores a value in a variable. Typing `=` where you need `==` is one of the most common mistakes in Python.
- Both the `if` line and the `else` line end with a colon `:`.
- The lines under `if` and under `else` are indented by four spaces, just like the lines inside a loop.

**Steps:**
1. Define a secret launch code.
2. Ask the operator to type the code.
3. If the code is correct, the drone flies. Otherwise, it stays grounded.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
# The secret launch code. In real systems, a password should never be shared.
secret_code = "SQUADRONE2026"

# Ask the operator to enter the code
attempt = input("Enter the launch code to unlock the drone: ")

if attempt == secret_code:
    print("Access granted. Launching.")
    tello.takeoff()
    tello.rotate_clockwise(360)
    tello.land()
else:
    print("Access denied. Incorrect code - the drone will not fly.")

print("Mission 10 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 11: Writing a Function

When the same routine is flown repeatedly, you can give it a name and reuse it. This is what a **function** is - a named set of instructions that you can run whenever you need it.

A function is created with the keyword `def` (short for "define"):

```python
def my_function():
    # instructions go here, indented
```

Once defined, it is run by writing its name followed by brackets, such as `my_function()`.

In this mission we place the Hello World routine into a function named `hello_drone`, then call it.

**Steps:**
1. Define a function named `hello_drone` that takes off, climbs, rotates, and lands.
2. Call the function to run it.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
def hello_drone():
    tello.takeoff()
    tello.move_up(80)
    tello.rotate_counter_clockwise(360)
    tello.land()

# Run the function
hello_drone()

print("Mission 11 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 12: Functions with Parameters

Functions become more powerful when you give them an **input value**, called a **parameter**. This lets one function behave differently each time it is called.

For example, a `fly_square` function can take a `size` parameter, so the same code can fly a small or large square:

```python
def fly_square(size):
    ...
```

Then `fly_square(50)` flies a small square and `fly_square(150)` flies a large one.

**Steps:**
1. Define a function `fly_square(size)` that uses a loop to fly a square of the given size.
2. Call it with a small size.
3. Optionally, call it again with a larger size.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
def fly_square(size):
    tello.takeoff()
    for i in range(4):
        tello.move_forward(size)
        tello.rotate_clockwise(90)
    tello.land()

# Fly a small square (80 cm sides)
fly_square(80)

# Try this as well - remove the # to fly a larger square:
# fly_square(150)

print("Mission 12 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 13: Combining Functions

Larger programs are built from small functions joined together. In this mission you will write three small functions and call them in order to form a complete routine.

**Steps:**
1. Write a function to take off and climb to a safe height.
2. Write a function that rotates.
3. Write a function that performs a flip.
4. Call all three in order, then land.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
def take_off_safely():
    tello.takeoff()
    tello.move_up(70)

def spin():
    tello.rotate_clockwise(360)

def do_a_flip():
    tello.flip_forward()

# Build the routine by calling the functions in order
take_off_safely()
spin()
do_a_flip()
tello.land()

print("Mission 13 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 14: Decisions Based on Sensors

A well-designed program checks that it is safe to fly before taking off. Flying with a low battery can cause the drone to fall, so the program should decide for itself.

This mission reads the battery sensor and uses an `if` statement: if the battery is high enough, the drone flies; otherwise it stays grounded and warns the operator. This combines sensors (Mission 9) with decisions (Mission 10).

**Steps:**
1. Read the battery level.
2. If the battery is above 30 percent, fly. Otherwise, print a warning and do not fly.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
battery = tello.get_battery()
print("Battery level is", battery, "%")

if battery > 30:
    print("Battery is sufficient - cleared for take-off.")
    tello.takeoff()
    tello.move_up(80)
    tello.land()
else:
    print("Battery is too low. It is not safe to fly. Please charge the drone.")

print("Mission 14 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Mission 15: A Launch Countdown

Launches are often preceded by a countdown. This can be built using a loop, the `time.sleep()` command (which pauses the program), and a `range` that counts **backwards**.

`range(5, 0, -1)` counts 5, 4, 3, 2, 1. The third number, `-1`, means decrease by one each step.

**Steps:**
1. Print a preparation message.
2. Loop from 5 down to 1, printing each number and pausing one second.
3. Launch.

## Mission 19: Handling Errors Safely

A reliable program plans for things going wrong. A command can sometimes fail - for example, if the battery drops or the connection is interrupted. The flight can be protected using `try` and `except`.

- The code inside `try` is what should happen normally.
- If anything goes wrong, the code inside `except` runs instead - and we use it to land the drone safely regardless of the error.
- The emergency landing itself sits inside its own small `try`, because if the problem was a lost connection, even the landing command can fail. In that case the program prints a warning and the drone lands itself when its safety systems take over.

**Steps:**
1. Place the flight inside a `try` block.
2. In the `except` block, print the problem and land safely.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import time

tello = Tello()
tello.connect()
print("Battery level:", tello.get_battery(), "%")

# --- MISSION ---
try:
    tello.takeoff()
    tello.move_up(80)
    tello.flip_forward()
    tello.land()
except Exception as e:
    print("Something went wrong:", e)
    print("Running emergency landing.")
    try:
        tello.land()
    except Exception:
        print("Could not reach the drone. Ask an instructor for help.")

print("Mission 19 complete.")

# --- POST-FLIGHT: release the drone for the next pilot ---
tello.end()
print("Drone released - the next pilot can now connect.")
````

## Bonus Mission: Build a Drone Control Panel

*Contributed by Academy pilot Rex Andrew Bennetts, and improved together during code review.*

Every mission so far has been a script: the code runs from top to bottom and the flight is over. This mission builds something different - an **application** with buttons you click to steer the drone live, using `tkinter`, the window toolkit that comes with Python (nothing extra to install).

### New ideas in this mission

- `tk.Tk()` creates a window. `.title(...)` names it and `.geometry("700x450")` sets its starting size in pixels.
- `tk.Button(mainwin, text="Land", command=land)` creates a clickable button. Look closely at `command=land` - there are **no parentheses** after the function name. You are not running the function; you are handing the function itself to the button, and `tkinter` runs it each time the button is clicked. We use the same trick to hand drone commands like `tello.move_forward` to `make_button`.
- `.place(relx=0.5, rely=0.35, anchor="center")` positions a button using **fractions of the window**: `relx=0.5` means halfway across, `rely=0.35` means about a third of the way down. Because the positions are fractions, the layout adjusts itself when the window is resized.
- `make_button` is a **function that builds a function**: each time it runs, it creates a fresh `press` function wired to one specific drone command, then attaches it to a new button. This is the Mission 11 and 12 idea taken one step further.
- `mainwin.mainloop()` starts the window and keeps it running until you close it. While the window is open, the notebook cell stays busy - that is normal.
- The window will **freeze for a moment during each move**: drone commands wait until the drone has finished moving before they return. That is also normal.

### Steps

1. Start with the usual pre-flight block, but store the battery in a variable first: `battery = tello.get_battery()`, then print it. The status bar uses this variable.
2. Create the window, set its title and size, and add a status `Label` at the top showing the battery level.
3. Write `make_button(label, action, value, relx, rely)`. Inside it, define `press()`, which runs the command inside `try`/`except` (Mission 19): commands like `takeoff` take no number, commands like `move_forward` take one, and any failure should appear in the status bar instead of crashing the panel.
4. Create the buttons using this flight deck plan:

| Label | Drone command | Value | relx | rely |
|---|---|---|---|---|
| Take off | `tello.takeoff` | `None` | 0.82 | 0.35 |
| Land | `tello.land` | `None` | 0.82 | 0.55 |
| Forward | `tello.move_forward` | `MOVE_DISTANCE` | 0.50 | 0.35 |
| Back | `tello.move_back` | `MOVE_DISTANCE` | 0.50 | 0.55 |
| Left | `tello.move_left` | `MOVE_DISTANCE` | 0.38 | 0.45 |
| Right | `tello.move_right` | `MOVE_DISTANCE` | 0.62 | 0.45 |
| Turn left | `tello.rotate_counter_clockwise` | `ROTATE_ANGLE` | 0.32 | 0.35 |
| Turn right | `tello.rotate_clockwise` | `ROTATE_ANGLE` | 0.68 | 0.35 |
| Rise | `tello.move_up` | `CLIMB_DISTANCE` | 0.18 | 0.35 |
| Lower | `tello.move_down` | `CLIMB_DISTANCE` | 0.18 | 0.55 |

5. Add a **Disconnect** button that runs `tello.end()` and updates the status bar. In this mission, that button *is* your post-flight block - press it before the next pilot connects.
6. Start the window with `mainwin.mainloop()`.

> **SAFETY:** the Take off button launches the drone the instant it is clicked. Check the space is clear *before* opening the panel, and keep your pointer near the Land button while flying. If you close the window without pressing Disconnect, run the **Emergency: Land and Release** cell at the bottom of the notebook.

````python
# --- PRE-FLIGHT: connect to the drone ---
from djitellopy import Tello
import tkinter as tk

tello = Tello()
tello.connect()
battery = tello.get_battery()
print("Battery level:", battery, "%")

# --- MISSION ---
MOVE_DISTANCE = 60     # cm per button press
CLIMB_DISTANCE = 50    # cm per Rise / Lower press
ROTATE_ANGLE = 45      # degrees per turn

mainwin = tk.Tk()
mainwin.title("Squadrone Control Panel")
mainwin.geometry("700x450")

# str(battery) turns the number into text so it can be joined to the message
status = tk.Label(mainwin, text="Connected. Battery: " + str(battery) + "%", font=("Arial", 12))
status.place(relx=0.5, rely=0.12, anchor="center")

def make_button(label, action, value, relx, rely):
    # Builds one button. Each click runs the drone command safely
    # and reports the result in the status bar.
    def press():
        try:
            if value is None:
                action()           # commands like takeoff() take no number
            else:
                action(value)      # commands like move_forward(60) take one
            status.config(text="Ready. Battery: " + str(tello.get_battery()) + "%")
        except Exception as error:
            status.config(text="Command failed: " + str(error))
    button = tk.Button(mainwin, text=label, font=("Arial", 14), width=9, command=press)
    button.place(relx=relx, rely=rely, anchor="center")

# label, drone command (no parentheses), value, horizontal position, vertical position
make_button("Take off", tello.takeoff, None, 0.82, 0.35)
make_button("Land", tello.land, None, 0.82, 0.55)
make_button("Forward", tello.move_forward, MOVE_DISTANCE, 0.50, 0.35)
make_button("Back", tello.move_back, MOVE_DISTANCE, 0.50, 0.55)
make_button("Left", tello.move_left, MOVE_DISTANCE, 0.38, 0.45)
make_button("Right", tello.move_right, MOVE_DISTANCE, 0.62, 0.45)
make_button("Turn left", tello.rotate_counter_clockwise, ROTATE_ANGLE, 0.32, 0.35)
make_button("Turn right", tello.rotate_clockwise, ROTATE_ANGLE, 0.68, 0.35)
make_button("Rise", tello.move_up, CLIMB_DISTANCE, 0.18, 0.35)
make_button("Lower", tello.move_down, CLIMB_DISTANCE, 0.18, 0.55)

# --- POST-FLIGHT: in this mission, the Disconnect button releases the drone ---
def disconnect():
    tello.end()
    status.config(text="Drone released - the next pilot can now connect.")

disconnect_button = tk.Button(mainwin, text="Disconnect", font=("Arial", 14), width=12, command=disconnect)
disconnect_button.place(relx=0.5, rely=0.8, anchor="center")

mainwin.mainloop()
print("Control panel closed.")
````
