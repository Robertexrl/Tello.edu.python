Mission 1 Fly Hello tello mission, take off, go up 100cm, turn 360 degrees (we pretend the drone says hello), and then we land.

This is an explanation of the first part of the code, it does look complex but don't worry, you will learn it with time:
Import Modules:

import socket: This brings in the socket module, which lets the program send and receive data over a network.
import threading: This is used to run code in different threads at the same time, like doing two things at once.
import time: This is used to make the program wait for a certain amount of time.
Set Up Addresses:

tello_address = ('192.168.10.1', 8889): This is the address of the Tello drone. It's like the drone's home address on the internet.
local_address = ('', 9000): This is the address for your computer. It's like saying, "I'll be waiting at door number 9000 to talk to the drone."
Create a UDP Connection:

sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM): This creates a new way for your computer to talk to other devices using UDP, which is a fast but simple type of internet communication.
Bind to Local Address:

sock.bind(local_address): This tells your computer to get ready to send and receive messages at door number 9000.
Send Message Function:

def send(message, delay): This is a set of instructions that lets you send a message to the drone.
Inside the send function:
It tries to send a message to the drone. If it can't, it will tell you there was an error.
time.sleep(delay): After sending the message, it waits for a bit (this is the delay).
Receive Message Function:

def receive(): This set of instructions listens for messages from the drone.
Inside the receive function:
It keeps checking for messages from the drone. If it gets a message, it shows you what the message says.
If there's a problem and it can't listen anymore, it will stop trying and tell you there was an error.
Listening Thread:

receiveThread = threading.Thread(target=receive): This creates a special task that can listen for drone messages while the rest of the program does other things.
receiveThread.daemon = True: This makes sure that this special task doesn't keep running forever if the main program stops.
receiveThread.start(): This starts the special task, so it begins listening for messages from the drone.
That's it! This code sets up a way for your computer to send instructions to a Tello drone and listen to what the drone says back, all while being able to do other things at the same time.

```python
# Import the necessary modules
import socket
import threading
import time

# IP and port of Tello
tello_address = ('192.168.10.1', 8889)

# IP and port of local computer
local_address = ('', 9000)

# Create a UDP connection that we'll send the command to
sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# Bind to the local address and port
sock.bind(local_address)

# Send the message to Tello and allow for a delay in seconds
def send(message, delay):
  # Try to send the message otherwise print the exception
  try:
    sock.sendto(message.encode(), tello_address)
    print("Sending message: " + message)
  except Exception as e:
    print("Error sending: " + str(e))

  # Delay for a user-defined period of time
  time.sleep(delay)

# Receive the message from Tello
def receive():
  # Continuously loop and listen for incoming messages
  while True:
    # Try to receive the message otherwise print the exception
    try:
      response, ip_address = sock.recvfrom(128)
      print("Received message: " + response.decode(encoding='utf-8'))
    except Exception as e:
      # If there's an error close the socket and break out of the loop
      sock.close()
      print("Error receiving: " + str(e))
      break

# Create and start a listening thread that runs in the background
# This utilizes our receive functions and will continuously monitor for incoming messages
receiveThread = threading.Thread(target=receive)
receiveThread.daemon = True
receiveThread.start()

#Put Tello into command mode
send("command", 3)

# Send the takeoff command
send("takeoff", 5)

# Go up by 100 cm
send("up 100", 4)

# Perform a 360-degree clockwise yaw
send("cw 360", 5)

# Land
send("land", 5)

# Print message
print("Mission completed successfully!")

# Close the socket
sock.close()

 
```

Mission 4 Fly a square and repeat 3 times (short way). Every time a square is completed, go up 100cm and start a new square

Explaining Loops:
Imagine you have a favorite song, and you want to listen to it three times in a row. You could hit the 'play' button, wait for the song to finish, and then hit 'play' again two more times. But what if you had a magic button that you could tell, "Play my favorite song three times," and it would do it for you without you needing to press 'play' again and again? That's kind of what a loop in programming does! so we will use here!

```python
# Import the necessary modules
import socket
import threading
import time

# IP and port of Tello
tello_address = ('192.168.10.1', 8889)

# IP and port of local computer
local_address = ('', 9000)

# Create a UDP connection that we'll send the command to
sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# Bind to the local address and port
sock.bind(local_address)

# Send the message to Tello and allow for a delay in seconds
def send(message, delay):
    # Try to send the message otherwise print the exception
    try:
        sock.sendto(message.encode(), tello_address)
        print("Sending message: " + message)
    except Exception as e:
        print("Error sending: " + str(e))

    # Delay for a user-defined period of time
    time.sleep(delay)

# Receive the message from Tello
def receive():
    # Continuously loop and listen for incoming messages
    while True:
        # Try to receive the message otherwise print the exception
        try:
            response, ip_address = sock.recvfrom(128)
            print("Received message: " + response.decode(encoding='utf-8'))
        except Exception as e:
            # If there's an error close the socket and break out of the loop
            sock.close()
            print("Error receiving: " + str(e))
            break

# Create and start a listening thread that runs in the background
# This utilizes our receive functions and will continuously monitor for incoming messages
receiveThread = threading.Thread(target=receive)
receiveThread.daemon = True
receiveThread.start()

# Put Tello into command mode
send("command", 3)

# Send the takeoff command
send("takeoff", 5)

# Each leg of the box will be 100 cm. Tello uses cm units by default.
box_leg_distance = 100

# Yaw 90 degrees
yaw_angle = 90

# Yaw clockwise (right)
yaw_direction = "cw"

# Number of squares to fly
num_squares = 3

for _ in range(num_squares):
    # Fly a square
    for _ in range(4):
        send("forward " + str(box_leg_distance), 4)
        send(yaw_direction + " " + str(yaw_angle), 3)

    # Go up by 100 cm after completing a square
    send("up 100", 4)

# Land
send("land", 5)

# Print message
print("Mission completed successfully!")

# Close the socket
sock.close()
```

Explaining the Specific loop code above:

This code is like telling a drone to fly in a square shape, and then go up a little, and do that whole thing a few times (as many as num_squares says).

for _ in range(num_squares): - This is like saying, "Do the following steps a certain number of times." num_squares is how many times you want to do it. If num_squares is 3, it's like saying, "Do the next steps 3 times."

Inside the first loop, there's another loop:

for _ in range(4): - This is saying, "Do the next steps 4 times." This is for flying in a square because a square has four sides.
Inside this loop, there are two commands: send("forward " + str(box_leg_distance), 4) and send(yaw_direction + " " + str(yaw_angle), 3). These are like telling the drone, "Go forward a certain distance" and then "Turn a corner." Doing this four times makes the drone fly a square.
send("up 100", 4) - After flying a square, this tells the drone, "Now go up 100 cm." This happens each time after the drone completes a square.

So, the drone flies a square, goes up, flies another square a little higher, goes up again, and keeps doing this until it has done it the number of times we told it to in num_squares.

Mission 6 fly a perfect circle around a focal point, this is our introduction to photogametry and then let's expand the code and make it into a drone choreography! 

Story: you have been contracted by an oil and gas company to create a 3D model of their platform. They have calculated that you need to take a total of 12 photographs around it, so you need to fly in a circle with the camera of the drone always facing the platform, the distance from the oil and gas platform to the photo point is 200cm. 

We will let you think about it and start planning how you can do this. There are many ways to do it and you can try your own.

Once you complete your circle, we will turn this into a choreography for a mini-drone show, the whole class will need to collaborate to plan a drone choreography.
First the whole class needs to come up with a concept and plan what you want to do. Then each team will be responsible to for correctly programing your own drone.
Make sure you think ahead of time to ensure your drones don't collide with each other. 

```python
import socket
import threading
import time
import math

# IP and port of Tello
tello_address = ('192.168.10.1', 8889)

# IP and port of local computer
local_address = ('', 9000)

# Create a UDP connection that we'll send the command to
sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# Bind to the local address and port
sock.bind(local_address)

# Send the message to Tello and allow for a delay in seconds
def send(message, delay):
    # Try to send the message otherwise print the exception
    try:
        sock.sendto(message.encode(), tello_address)
        print("Sending message: " + message)
    except Exception as e:
        print("Error sending: " + str(e))

    # Delay for a user-defined period of time
    time.sleep(delay)

# Receive the message from Tello
def receive():
    # Continuously loop and listen for incoming messages
    while True:
        # Try to receive the message otherwise print the exception
        try:
            response, ip_address = sock.recvfrom(128)
            print("Received message: " + response.decode(encoding='utf-8'))
        except Exception as e:
            # If there's an error close the socket and break out of the loop
            sock.close()
            print("Error receiving: " + str(e))
            break

# Create and start a listening thread that runs in the background
receiveThread = threading.Thread(target=receive)
receiveThread.daemon = True
receiveThread.start()

# Put Tello into command mode
send("command", 3)

# Send the takeoff command
send("takeoff", 5)

# Define the parameters for the flight path
radius = 200  # Radius in cm
photo_count = 12
angle_increment = 360 / photo_count

# Fly up to the initial height
send(f"up 100", 5)

# Fly forward to start the circle
send(f"forward {radius}", 5)

# Rotate 180 degrees to face the center of the circle
send("cw 180", 3)

# Fly the circle in segments
for count in range(photo_count):
    # Hover for 3 seconds (pause the script)
    time.sleep(3)

    # Rotate by the angle increment
    send(f"cw {angle_increment}", 3)

    # Calculate the distance to fly for the next segment
    x1 = radius * math.cos(math.radians(angle_increment * count))
    y1 = radius * math.sin(math.radians(angle_increment * count))
    x2 = radius * math.cos(math.radians(angle_increment * (count + 1)))
    y2 = radius * math.sin(math.radians(angle_increment * (count + 1)))
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

    # Fly the calculated distance
    send(f"right {int(distance)}", 5)

# Land the drone
send("land", 5)
print("Mission completed successfully!")
sock.close()
```
